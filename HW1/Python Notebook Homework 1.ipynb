```python
#imports
import mediacloud, json, datetime

#you need to add your mediacloud key. Available here https://core.mediacloud.org/login/register
mc = mediacloud.api.MediaCloud('MY_API_KEY')

#request based on Media Cloud Python API examples here https://github.com/c4fcm/MediaCloud-API-Client
fetch_size = 2000
stories = []
last_processed_stories_id = 0
while len(stories) < 2000:
    fetched_stories = mc.storyList('( hillary AND clinton ) OR ( donald AND trump)', #specified first names to avoid other references
                                   solr_filter=[ mc.publish_date_query( datetime.date(2016, 10, 8), datetime.date(2016, 10, 11)), #modified dates to track changes in media as a result of Trump's tape being leaked
                                                                         'tags_id_media:1'],
                                    last_processed_stories_id=last_processed_stories_id, rows= fetch_size)
    stories.extend(fetched_stories)
    if len(fetched_stories) < fetch_size:
        break

    last_processed_stories_id = stories[-1]['processed_stories_id']

print json.dumps(stories)
```
